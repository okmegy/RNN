```python
import csv
import itertools
import operator
import numpy as np
import nltk
import sys
from datetime import datetime
```

```python
def softmax(x):
    xt = np.exp(x - np.max(x))
    return xt / np.sum(xt)
```

```python
vocabulary_size = 8000  
unknown_token = "UNKNOWN_TOKEN"  
sentence_start_token = "SENTENCE_START"  
sentence_end_token = "SENTENCE_END"
```

```python
Project_path = "/home/jasper/python-notebook"
sys.path.append(Project_path)
```

```python
with open('reddit-comments-2015-08.csv', 'rb') as f:
    reader = csv.reader(f, skipinitialspace=True)
    reader.next()
    sentences = itertools.chain(*[nltk.sent_tokenize(x[0].decode('utf-8').lower()) for x in reader])
    sentences = ["%s %s %s" % (sentence_start_token, x, sentence_end_token) for x in sentences]
    print "Parsed %d sentences." % (len(sentences))
```

```python
#separate the sentence into word
tokenized_sentences = [nltk.word_tokenize(sent) for sent in sentences]
```

```python
#count frequency 
word_freq = nltk.FreqDist(itertools.chain(*tokenized_sentences))  
print "Found %d unique words tokens." % len(word_freq.items())
```

```python
vocab = word_freq.most_common(vocabulary_size-1)  
```

```python
index_to_word = [x[0] for x in vocab] 
```

```python
index_to_word.append(unknown_token)
```

```python
word_to_index = dict([(w,i) for i,w in enumerate(index_to_word)])
```

```python
print "Using vocabulary size %d." % vocabulary_size
```

```python
print "The least frequent word in our vocabulary is '%s' and appeared %d times." % (vocab[-1][0], vocab[-1][1])
```

```python
X_train  = []
for sen in tokenized_sentences:
    tmp = []
    for i in sen[:-1]:
        if word_to_index.get(i)==None:
            continue
        tmp.append(word_to_index.get(i))
    X_train.append(tmp)
```

```python
Y_train  = []
for sen in tokenized_sentences:
    tmp = []
    for i in sen[1:]:
        if word_to_index.get(i)==None:
            continue
        tmp.append(word_to_index.get(i))
    Y_train.append(tmp)
```

```python
class RNNNumpy:
    def __init__(self,word_dim,hidden_dim,bptt_truncate=4):
        self.word_dim = word_dim
        self.hidden_dim = hidden_dim
        self.bptt_truncate = bptt_truncate
        self.U = np.random.uniform(-np.sqrt(1./word_dim), np.sqrt(1./word_dim), (hidden_dim, word_dim))
        self.V = np.random.uniform(-np.sqrt(1./hidden_dim), np.sqrt(1./hidden_dim), (word_dim, hidden_dim))
        self.W = np.random.uniform(-np.sqrt(1./hidden_dim), np.sqrt(1./hidden_dim), (hidden_dim, hidden_dim))
        
    def forward_propagation(self,x):
        T = len(x)
        s = np.zeros((T+1,self.hidden_dim))
        s[-1] = np.zeros(self.hidden_dim)
        o = np.zeros((T,self.word_dim))
        for t in np.arange(T):
            s[t] = np.tanh(self.U[:,x[t]] + self.W.dot(s[t-1])) #pick up the first k column + multiply s[t-1] by W
            o[t] = softmax(self.V.dot(s[t]))
        return [o,s]
    
    def predict(self,x):
        o, s = self.forward_propagation(x)
        return np.argmax(o, axis=1)
```

```python
test = RNNNumpy(8000,100)
```
